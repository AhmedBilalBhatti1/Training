```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    @classmethod
    def from_birth_year(cls, name, birth_year):
        return cls(name, 2026 - birth_year)

p1 = Person.from_birth_year("Charlie", 2000)
print(p1.age)
```

Polymorphism

```python
class Animal:
    def sound(self):
        print("Some sound")

class Dog(Animal):
    def sound(self):
        print("Bark")

class Cat(Animal):
    def sound(self):
        print("Meow")

animals = [Dog(), Cat()]

for animal in animals:
    animal.sound()
```

```python
class Person:
    species = "Human"

    def __init__(self, name):
        self.name = name  

    def greet(self):   
        print("Hi, I am", self.name)

    @classmethod
    def get_species(cls):
        print("Species:", cls.species)

    @staticmethod
    def is_adult(age):
        return age >= 18


p = Person("Alice")

p.greet()        
Person.get_species()   
print(Person.is_adult(20)) 
```

```python
class CountUp:
    def __init__(self, max):
        self.max = max
        self.current = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.current > self.max:
            raise StopIteration
        value = self.current
        self.current += 1
        return value


counter = CountUp(3)

for num in counter:
    print(num)
```

```python
def count_up(max):
    current = 1
    while current <= max:
        yield current
        current += 1


counter = count_up(3)

for num in counter:
    print(num)
```
