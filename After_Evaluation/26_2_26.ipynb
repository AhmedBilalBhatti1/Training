```python
class Resource:
    def __new__(cls, name: str) -> "Resource":
        print(f"__new__ called for {name}")
        return super().__new__(cls)      

    def __init__(self, name: str) -> None:
        self.name = name
        print(f"__init__ called → Resource {name} created")

    def use(self) -> None:
        print(f"Using {self.name}")

    def __del__(self) -> None:
        print(f"__del__ called → Resource {self.name} destroyed")

# Dry-run
r1 = Resource("Database Connection")
r1.use()
del r1
```

```python
class Player:
    game_name: str = "Cricket"          # class variable

    def __init__(self, name: str) -> None:
        self.name = name                # instance variable
        self.score: int = 0

    @classmethod
    def change_game(cls, new_game: str) -> None:
        cls.game_name = new_game

    @staticmethod
    def is_valid_score(score: int) -> bool:   # no self/cls
        return score >= 0
```

```python
class BankAccount:
    bank_name = "HBL"
    total_accounts = 0

    def __init__(self, name, balance):
        self.name = name
        self.balance = balance
        BankAccount.total_accounts += 1

    # Instance Method
    def deposit(self, amount):
        self.balance += amount

    # Class Method
    @classmethod
    def change_bank(cls, new_name):
        cls.bank_name = new_name

    # Static Method
    @staticmethod
    def validate_amount(amount):
        return amount > 0
```

```python
a1 = BankAccount("Ali", 1000)
a2 = BankAccount("Sara", 2000)

a1.deposit(500)           # affects only Ali

BankAccount.change_bank("UBL")   # affects all accounts

print(BankAccount.validate_amount(200))  # helper logic
```

```python
from typing import overload

class Calculator:
    @overload
    def add(self, a: int, b: int) -> int: ...
    @overload
    def add(self, a: float, b: float) -> float: ...
    @overload
    def add(self, a: str, b: str) -> str: ...

    def add(self, a: int | float | str, b: int | float | str) -> int | float | str:
        return a + b


C = Calculator()
print(C.add(1, 2))
print(C.add(1.1, 2.2))
print(C.add("Hello ", "World"))
```

Custom Iterators

```python
from typing_extensions import Self

class Countdown:
    def __init__(self, start: int) -> None:
        self.current = start

    def decrement(self) -> Self:
        self.current -= 1
        return self

    def increment(self) -> Self:
        self.current += 1
        return self

a = Countdown(5).decrement().decrement().decrement().decrement().decrement().decrement()
print(a.current)
b = Countdown(5).increment().increment().increment().increment().increment().increment()
print(b.current)
```

Generators

```python
def countdown(n: int):
    while n > 0:
        yield n
        n -= 1

gen = countdown(5)
print(list(gen))
```

```python
def read_lines(path: str):
    with open(path, encoding="utf-8") as f:
        for line in f:         
            yield line.strip()

for line in read_lines("very_big_log.txt"):
    if "ERROR" in line:
        print(line)
```

Decorators

```python
def repeat(times):
    def wrapper(func):
        def inner(name):
            for _ in range(times):
                func(name)
        return inner
    return wrapper

@repeat(3)
def say_hello(name: str) -> None:
    print(f"Hello {name}!")

say_hello("Ahmed")
```

```python
from datetime import datetime

def add_timestamp(cls):
    old = cls.__init__

    def init(self, *a, **k):
        old(self, *a, **k)
        self.created_at = datetime.now()

    cls.__init__ = init
    return cls

@add_timestamp
class User:
    def __init__(self, name):
        self.name = name

user = User("Ahmed")
print(user.created_at)
```
